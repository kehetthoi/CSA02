```python
import pandas as pd
```

```python
df = pd.read_excel('provinces.xlsx', sheet_name='Sheet1')
df
```

```python
import matplotlib.pyplot as plt
```

```python
df_select = df.head(10)

plt.figure(figsize=(10,6))
plt.bar(df_select['Name'],df_select['Population'],color='skyblue')
plt.title('Dan so 10 tinh dau tien',fontsize=14)
plt.xlabel('Tinh/thanh pho',fontsize=12)
plt.ylabel('Dan so',fontsize=12)
plt.xticks(rotation=90)
plt.show()
```

```python
df_region = df.groupby('Region').agg({
 'Area':'sum',
 'Population':'sum'
}).reset_index()

df_region['Density'] = df_region['Population'] / df_region['Area']
df_region
```

```python
df_dense=df.groupby('Region')[['Population','Area']].sum().reset_index()
df_dense['Density'] = df_dense['Population'] / df_dense['Area']
df_dense
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
plt.bar(df_dense['Region'], df_dense['Density'], color='skyblue')
plt.title('Mật độ dân số theo vùng', fontsize=14)
plt.xlabel('Vùng', fontsize=12)
plt.ylabel('Mật độ dân số', fontsize=12)
plt.xticks(rotation=90, ha='right') 
plt.tight_layout()
plt.show()
```

```python

import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
plt.bar(df_region['Region'], df_region['Density'], color='lightgreen')
plt.title('Mật độ dân số theo vùng', fontsize=14)
plt.xlabel('Vùng', fontsize=12)
plt.ylabel('Mật độ dân số', fontsize=12)
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()  

```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
plt.plot(df_region['Region'], df_region['Density'],marker='o', color='lightgreen',linewidth=2)
plt.title('Mật độ dân số theo vùng', fontsize=14)
plt.xlabel('Vùng', fontsize=12)
plt.ylabel('Mật độ dân số', fontsize=12)
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()  
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 10))
plt.pie(df_region['Density'], labels=df_region['Region'],autopct='%1.1f%%', startangle=140,colors=plt.cm.Paired.colors)
plt.title('Mật độ dân số theo vùng', fontsize=14)
plt.axis('equal') 
plt.show()
```
