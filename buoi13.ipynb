```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('provinces_updated.xlsx', sheet_name='Sheet1')
df
```

```python
import matplotlib.pyplot as plt
df_sorted = df.sort_values('Population')
plt.figure(figsize=(16, 10)) 
plt.bar(df_sorted['Name'], df_sorted['Population'], color='skyblue')
plt.title('Dân số theo thứ tự tăng dần', fontsize=16)
plt.xlabel('Tên thành phố', fontsize=13)
plt.ylabel('Dân số', fontsize=13)
plt.xticks(rotation=45, ha='right', fontsize=11) 
plt.tight_layout()
plt.show()

```

```python
df=pd.read_excel('region_areas.xls')
df
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 10))  
plt.pie(
    df['Total Area (km2)'], 
    labels=df['Region'], 
    autopct='%1.1f%%', 
    startangle=140, 
    colors=plt.cm.Paired.colors
)
plt.title('Mật độ dân số theo vùng', fontsize=14)
plt.axis('equal')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_excel('provinces_updated.xlsx')
df['Frequency'] = df['Area'] / df['Population']
df
```

```python
df_sorted = df.sort_values(by='Density', ascending=True).reset_index()
df_sorted
```

```python
plt.figure(figsize=(15, 5))
plt.bar(df_sorted['Density'], df_sorted['Frequency'], color='skyblue')
plt.xlabel('Population Density')
plt.ylabel('Frequency')
plt.title("Population's frequency and density")
plt.tight_layout()
plt.show()
```
